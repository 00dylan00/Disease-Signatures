```python
import pandas as pd
import requests

# Endpoint for retrieving analyses
BASE_URL = 'http://stargeo.org/api/v2/analysis/'

# Fetch the first page of results
data = requests.get('http://stargeo.org/api/v2/analysis/').json()

# Extract the results from the first page
analysis = data['results'][0]

# Fetch source dataframe
analysis_df = pd.read_json(analysis['df'], orient='split')
analysis_df.head()

# # Fetch fold changes. WARNING: this could be big
# r = requests.get(analysis['fold_changes'])

# # It is also compressed with zlib
# import zlib
# fold_changes = pd.read_json(zlib.decompress(r.content), orient='split')
# fold_changes.head()
```

```python
import requests

base_url = 'http://stargeo.org/api/v2/analysis/'
results = []

while base_url:
    data = requests.get(base_url).json()
    results.extend(data['results'])
    base_url = data['next']

# Now, `results` contains all the analyses. Convert them to a DataFrame:
df = pd.DataFrame(results)
```

```python
filtered_df = df[(df["specie"] == "human") & (df["df"].notnull())]
```
